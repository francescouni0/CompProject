```python
import os
import numpy as np
import scipy
import reading 
from feature_extractor import feature_extractor
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn import svm, metrics
from sklearn.feature_selection import RFECV
from sklearn.pipeline import Pipeline
```

Load dei Path, Con il sort giusto per far corrispondere i soggetti

```python
paths_FA= reading.data_path("Diffusion_parameters_maps-20230215T134959Z-001","corrected_FA_image")
paths_masks=reading.data_path("Diffusion_space_segmentations-20230215T134839Z-001","Diffusion_space_segmentations-20230215T134839Z-001")


#print((os.path.basename(paths_masks[0]).split('_')[2]))

paths_FA.sort(key=lambda x: int(os.path.basename(x).split('_')[3]))
paths_masks.sort(key=lambda x: int(os.path.basename(x).split('_')[2]))
```

Feature extraction

```python
region, mean, std=feature_extractor(paths_FA,paths_masks)
```

Reshape degli array per fare il Dataframe

```python
mean_t=np.transpose(mean)
std_t=np.transpose(std)
```

Creo DataFrame per mean and std

```python
df_mean = pd.DataFrame(mean_t[1:,1:177],index=mean[0][1:98],columns=region[1:177])
df_std=pd.DataFrame(std_t[1:,1:177],index=std[0][1:98],columns=region[1:177])

df_mean.head()
```

Aggiungo i gruppi

```python
a=pd.DataFrame(pd.read_csv('ADNI_dataset_diffusion.csv'))
a.sort_values(by=["Subject"],inplace=True)
group=a["Group"]
a.head()
```

SVM: creo i set e splitto

```python
X=df_mean.values
y=group.values

X_tr, X_tst, y_tr, y_tst = train_test_split(X, y, test_size=.2, random_state=27)
```

Modello gridsearch


```python
clf = svm.SVC(kernel="linear")

clf.fit(X_tr, y_tr)
y_pred = clf.predict(X_tst)
```

```python
print(metrics.classification_report(y_tst, y_pred)) 
```

Hyperparams tuning con Grid Search

```python
r=scipy.stats.expon.rvs(size=100)
g=scipy.stats.expon(scale=.1)
G=g.rvs(size=100)
# defining parameter range 
param_grid = {'C': r,
              'gamma': G, 
              'kernel': ['linear'], 
              'class_weight':['balanced', None]}

clf = svm.SVC(kernel="linear")

grid = GridSearchCV(clf, param_grid, refit = True,n_jobs=-1) 


# fitting the model for grid search 
grid.fit(X_tr, y_tr) 
 
# print best parameter after tuning 
print(grid.best_params_) 
grid_predictions = grid.predict(X_tst) 
print(grid_predictions)
   
# print classification report 
print(metrics.classification_report(y_tst, grid_predictions)) 
```

```python
#RFE on simple classifier

rfecv = RFECV(
    estimator=svm.SVC(kernel="linear"),
    step=1,
    cv=5,
    scoring="accuracy",
    min_features_to_select=len(a["Subject"]),
    n_jobs=-1,
)

rfecv.fit(X_tr, y_tr)
 
# print best parameter after tuning 
predictions = rfecv.predict(X_tst) 
print(predictions)
   
# print classification report 
print(metrics.classification_report(y_tst, predictions)) 
```

```python
#RFE on grid classifier (result of GridsearchCV)

rfecv = RFECV(
    estimator=grid.best_estimator_,
    step=1,
    cv=5,
    scoring="accuracy",
    min_features_to_select=len(a["Subject"]),
    n_jobs=-1,
)

rfecv.fit(X_tr, y_tr)
 
# print best parameter after tuning 
predictions = rfecv.predict(X_tst) 
print(predictions)
   
# print classification report 
print(metrics.classification_report(y_tst, predictions)) 
```

Hyperparams tuning con Grid Search e Feature reduction

```python
import time

start = time.time()

r=scipy.stats.expon.rvs(size=10)
g=scipy.stats.expon(scale=.1)
G=g.rvs(size=10)
# defining parameter range 
param_grid = {'estimator__C': r,
              'estimator__gamma': G, 
              'estimator__kernel': ['linear'], 
              'estimator__class_weight':['balanced', None]}

clf = svm.SVC(kernel="linear")

print("Checkpoint 0")

'''
rfecv_1 = RFECV(
    estimator=svm.SVC(kernel="linear"),
    step=1,
    cv=5,
    scoring="accuracy",
    min_features_to_select=len(a["Subject"]),
    n_jobs=-1,
)
'''

rfecv = RFECV(
    estimator=clf,
    step=1,
    cv=5,
    scoring="accuracy",
    min_features_to_select=len(a["Subject"]),
    n_jobs=-1,
)

print("Checkpoint 1")

grid = GridSearchCV(rfecv, param_grid, refit = True, n_jobs=-1) 

print("Checkpoint 2")

# fitting the model
grid.fit(X_tr, y_tr)

print("Checkpoint 3")
 
# print best parameter after tuning 
predictions = grid.predict(X_tst) 
print(predictions)
   
# print classification report 
print(metrics.classification_report(y_tst, predictions)) 

end = time.time()
print("Time: ", end-start)
```

Hyperparams tuning con Grid Search e Feature reduction implementati in una pipeline

```python
r=scipy.stats.expon.rvs(size=100)
g=scipy.stats.expon(scale=.1)
G=g.rvs(size=100)
# defining parameter range 
param_grid = {'model__C': r,
              'model__gamma': G, 
              'model__kernel': ['linear'], 
              'model__class_weight':['balanced', None]}

clf = svm.SVC(kernel="linear")

pipeline = Pipeline(steps = [("f_selection", RFECV(estimator=clf, step=1, cv=5, scoring="accuracy", min_features_to_select=len(a["Subject"]), n_jobs=-1)),
                             ("model", svm.SVC(kernel="linear"))])

grid = GridSearchCV(pipeline, param_grid, refit = True, n_jobs=-1) 

# fitting the model
grid.fit(X_tr, y_tr) 
 
# print best parameter after tuning 
predictions = grid.predict(X_tst) 
print(predictions)
   
# print classification report 
print(metrics.classification_report(y_tst, predictions)) 
```
